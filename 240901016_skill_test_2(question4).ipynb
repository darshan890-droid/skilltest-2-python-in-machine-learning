```python
import math

# 1. Euclidean Distance (2D)

x1, y1 = 2, 3
x2, y2 = 5, 7

distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print("Euclidean Distance:", distance)

# 2. Dot Product (Vectors)

A = [1, 2, 3]
B = [4, 5, 6]

dot_product = 0

for i in range(len(A)):
    dot_product += A[i] * B[i]

print("Dot Product:", dot_product)
```
