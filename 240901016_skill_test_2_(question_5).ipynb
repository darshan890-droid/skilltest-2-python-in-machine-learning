```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Create Date Range (1 year)
# -----------------------------
dates = pd.date_range(start="2025-01-01", periods=365)

# -----------------------------
# 2. Generate Random Temperatures using NumPy
# -----------------------------
np.random.seed(10)
temperature = np.random.randint(20, 45, size=365)

# -----------------------------
# 3. Create DataFrame
# -----------------------------
df = pd.DataFrame({"Date": dates, "Temperature": temperature})

# -----------------------------
# 4. Basic Analysis
# -----------------------------
print("First 5 rows:\n", df.head())

print("\nAverage Temperature:", df["Temperature"].mean())
print("Maximum Temperature:", df["Temperature"].max())
print("Minimum Temperature:", df["Temperature"].min())

# -----------------------------
# 5. Monthly Average Temperature
# -----------------------------
df["Month"] = df["Date"].dt.month
monthly_avg = df.groupby("Month")["Temperature"].mean()

print("\nMonthly Average Temperature:\n", monthly_avg)

# -----------------------------
# 6. Visualization
# -----------------------------

# Line plot (Time Series)
plt.figure(figsize=(10,5))
plt.plot(df["Date"], df["Temperature"], color="blue")
plt.title("Daily Temperature Over One Year")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()

# Bar plot (Monthly Average)
monthly_avg.plot(kind="bar", color="orange")
plt.title("Monthly Average Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```
