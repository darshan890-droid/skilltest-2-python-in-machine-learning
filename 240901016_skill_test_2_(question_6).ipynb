```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Create Date Range (1 year)
# -----------------------------
dates = pd.date_range(start="2025-01-01", periods=365)

# -----------------------------
# 2. Generate Stock Prices
# -----------------------------
np.random.seed(5)

price = 100  # starting price
prices = []

for i in range(365):
    change = np.random.normal(0, 1)   # small daily change
    price = price + change
    prices.append(price)

# -----------------------------
# 3. Create DataFrame
# -----------------------------
df = pd.DataFrame({"Date": dates, "Price": prices})

# -----------------------------
# 4. Basic Analysis
# -----------------------------
print("First 5 rows:\n", df.head())

print("\nAverage Price:", df["Price"].mean())
print("Maximum Price:", df["Price"].max())
print("Minimum Price:", df["Price"].min())

# -----------------------------
# 5. Moving Averages
# -----------------------------
df["MA_7"] = df["Price"].rolling(window=7).mean()
df["MA_30"] = df["Price"].rolling(window=30).mean()

print("\nMoving Averages:\n", df[["Price", "MA_7", "MA_30"]].head(10))

# -----------------------------
# 6. Visualization
# -----------------------------
plt.figure(figsize=(10,5))

plt.plot(df["Date"], df["Price"], label="Price", color="blue")
plt.plot(df["Date"], df["MA_7"], label="7-day MA", color="green")
plt.plot(df["Date"], df["MA_30"], label="30-day MA", color="red")

plt.title("Stock Price Analysis")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()

plt.show()
```
