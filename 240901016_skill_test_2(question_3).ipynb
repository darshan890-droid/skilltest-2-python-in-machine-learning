```python
import pandas as pd
import matplotlib.pyplot as plt

# Load file
df = pd.read_csv("flight_data.csv")

# a) Summary statistics for departure delays
print("Summary Statistics:\n")
print(df["DepartureDelay"].describe())

# b) Average delay for each airline
print("\nAverage Delay per Airline:\n")
print(df.groupby("Airline")["DepartureDelay"].mean())

# c) Most common origin and destination
print("\nMost Common Origin:")
print(df["Origin"].mode()[0])

print("\nMost Common Destination:")
print(df["Destination"].mode()[0])

# d) Box plot of departure delays
plt.boxplot(df["DepartureDelay"])
plt.title("Departure Delay Distribution")
plt.ylabel("Delay (minutes)")
plt.show()
```
